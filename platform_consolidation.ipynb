```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load your dataset
df = pd.read_csv('vgsales_clean.csv')  # Replace with your actual dataset

# Select relevant columns
df = df[['Name', 'Genre', 'Publisher', 'Year', 'Global_Sales', 'Decade', 'FranchiseTag']].dropna()
df = df.reset_index(drop=True)

# summary of df columns
print(df.info())
```

```python
# Consolidating rows with the same 'Name' and aggregating sales
df = df.groupby('Name').agg({
    'Genre': 'first',
    'Publisher': 'first',
    'Year': 'first',
    'Global_Sales': 'sum',
    'Decade': 'first',
    'FranchiseTag': 'first'
}).reset_index()
print(df.info())
```

```python
# Downloading dataset as CSV
df.to_csv('vgsales_consolidated.csv', index=False)
```
